```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = load_breast_cancer()
feature_names = data.feature_names
data_points = data.data
print(data.keys())
```

```python
df = pd.DataFrame(data_points, columns=feature_names)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_points, data.target, test_size=0.20, random_state=42)
```

```python
ada_clf = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1), n_estimators=200, algorithm="SAMME.R", learning_rate=0.5)
```

```python
ada_clf.fit(X_train, y_train)
```

```python
preds = ada_clf.predict(X_test)
```

```python
results = []
for x in range(10, 200, 10):
    ada_clf = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1), n_estimators=x, algorithm="SAMME.R", learning_rate=0.5)
    ada_clf.fit(X_train, y_train)
    preds = ada_clf.predict(X_test)   
    results.append([x, metrics.accuracy_score(y_test, preds), metrics.precision_score(y_test, preds), metrics.recall_score(y_test, preds)])    
```

```python
X =  [ x for x, _, _, _ in results ]
accuracy =  [ y for x, y, _, _ in results ]
precision =  [ y for x, _, y, _ in results ]
recall     = [ y for x, _, _, y in results  ]

font = {'family' : 'normal','weight' : 'bold','size'   : 22}
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(15,6))
plt.plot(X, accuracy)
plt.plot(X, precision)
plt.plot(X, recall)
plt.legend(['y = accuracy', 'y = precision', 'y = recall'], loc='upper left')
```
